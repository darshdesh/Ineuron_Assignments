Q2. Create the below pattern using nested for loop in Python. 
 
     *
     **
     ***
     ****
     *****
     ****
     ***
     **
     *


```python
# Answer:-
# We will use nested for loops for generating the pattern
for i in range(0,5):  #Outer Loop
    for j in range(0,1+i):  #Inner loop ascending order
        print("*",end="")   #Pattern element as '*'
    print()
for i in range(0,4):  #outer loop from bottom to up
    for k in range(0,4-i): #Inner loop with descending order
        print("*",end="")
    print()  #End of Program
```

Q2. Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line. 

```python
# Answer:-
r=[]  #Create a blank list
for i in range(2000,3201):      #Data Range from 2000 to 3200 including both
    if (i%7==0) and (i%5!=0):   # Condition Fulfillment as divisible by 7 but not divisible by 5 with assignment operator
        if i % 5 !=0:           # nested condition to add result in the list
            r.append(i)         # Appending the final data after condition fulfillment in the blank list 'r'
print(r)                        # Display the final output in comma seperated list
```

Q3. Write a Python program to accept the user's first and last name and then getting them printed in the the reverse order with a space between first name and last name. 

```python
#Answer:-
FN=input("Enter First Name: ")
LN=input("Enter Last Name: ")
Name= FN + " " + LN         #String Concatenate with space between them
print(Name[::-1])           # Using Slice operator to Reverse the Final String
```
